```python
from bs4 import BeautifulSoup
import pandas as pd

# Load the HTML content from the uploaded file
with open('Databanka testových úloh z českých reálií – Občanství.html', 'r', encoding='utf-8') as file:
    html_content = file.read()

# Parse the HTML content
soup = BeautifulSoup(html_content, 'html.parser')

# Initialize lists to store the data
questions = []
options = []
correct_answers = []

# Find all questions and answers
for question_block in soup.find_all('ol', class_='patnact ulohy_sub_1'):
    for question_item in question_block.find_all('li'):
        question_div = question_item.find('div', class_='text')
        if question_div:
            question_text = question_div.text.strip()
        else:
            continue
        
        answer_options = []
        for option in question_item.find_all('li'):
            label = option.find('label')
            if label:
                answer_options.append(label.text.strip())
        
        correct_answer_element = question_item.find('span', class_='spravne')
        if correct_answer_element:
            correct_answer = correct_answer_element.text.strip().replace('Správná odpověď: ', '')
        else:
            correct_answer = None
        
        questions.append(question_text)
        options.append(answer_options)
        correct_answers.append(correct_answer)

# Create a DataFrame
df = pd.DataFrame({
    'Question': questions,
    'Options': options,
    'Correct Answer': correct_answers
})

#import ace_tools as tools; tools.display_dataframe_to_user(name="Questions and Answers", dataframe=df)

df.head()
```
